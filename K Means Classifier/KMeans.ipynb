```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set() 
import numpy as np
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
iris = datasets.load_iris()
features= iris.data
labels = iris.target
```

```python
kmeans = KMeans(n_clusters=3, init='random', max_iter=500)
pred_y = kmeans.fit_predict(features)
print(pred_y)
print(labels)
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy :: ", accuracy_score(labels, pred_y, normalize=True))
```

```python
plt.scatter(features[:, 1], features[:, 2], c=pred_y, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 1], centers[:, 2], c='black', s=200, alpha=0.5);
```

```python
loss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(features)
    loss.append(kmeans.inertia_)
plt.plot(range(1, 11), loss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Loss')
plt.show()
```
